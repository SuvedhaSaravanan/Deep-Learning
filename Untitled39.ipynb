```python
class Device:
    def __init__(self, name, type_):
        self.name = name
        self.type_ = type_
        self.connections = []

    def add_connection(self, device):
        self.connections.append(device)

class Computer(Device):
    def __init__(self, name):
        super().__init__(name, "computer")

class Repeater(Device):
    def __init__(self, name):
        super().__init__(name, "repeater")

class Network:
    def __init__(self):
        self.devices = {}

    def add_device(self, device):
        self.devices[device.name] = device

    def add_connection(self, device1, device2):
        self.devices[device1].add_connection(device2)
        self.devices[device2].add_connection(device1)

    def find_route(self, start, end, route=None):
        if route is None:
            route = []
        route.append(start)
        if start == end:
            return route
        for device in self.devices[start].connections:
            if device not in route:
                new_route = self.find_route(device.name, end, route)
                if new_route:
                    return new_route
        return None

def main():
    network = Network()

    while True:
        print("1. Add a computer")
        print("2. Add a repeater")
        print("3. Add a connection")
        print("4. Print route")
        print("5. Quit")
        choice = int(input("Enter your choice: "))

        if choice == 1:
            name = input("Enter computer name: ")
            c = Computer(name)
            network.add_device(c)
            print(f"Computer {name} added to the network.")
        elif choice == 2:
            name = input("Enter repeater name: ")
            r = Repeater(name)
            network.add_device(r)
            print(f"Repeater {name} added to the network.")
        elif choice == 3:
            device1_name = input("Enter the name of the first device: ")
            device2_name = input("Enter the name of the second device: ")
            network.add_connection(device1_name, device2_name)
            print(f"Connection added between {device1_name} and {device2_name}.")
        elif choice == 4:
            start = input("Enter the start device name: ")
            end = input("Enter the end device name: ")
            route = network.find_route(start, end)
            if route:
                print("Route:", " -> ".join(route))
            else:
                print("No route found.")
        elif choice == 5:
            break
        else:
            print("Invalid choice.")

if __name__ == '__main__':
    main()
```
